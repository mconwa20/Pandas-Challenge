
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Gender Groupby

groupby_gender = purchase_data.groupby(["Gender"])

# Columns

purchase_count = groupby_gender["Gender"].count()

price_mean = groupby_gender["Price"].mean()

price_sum = groupby_gender["Price"].sum()

by_player = groupby_gender["SN"].nunique()

purchase_by_player = groupby_gender["Price"].sum()/groupby_gender["SN"].nunique()

# Print final dataframe

gender_summary = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price ($)": price_mean.round(2), 
                              "Total Purchase Value ($)": price_sum,
                              "Avg Total Purchase per Person ($)": purchase_by_player.round(2)})

gender_summary.head(20)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)

# Create age_count variable according to bins
age_count = purchase_data["Age Range"].value_counts()

#Groupby bins

groupby_age_range = purchase_data.groupby(["Age Range"])

# Create percentage of players variable

percentage_of_players = ((age_count/unique_SN) * 100).round(2)

age_demographics = pd.DataFrame({"Total Count": age_count,
                           "Percentage of Players (%)": percentage_of_players})

age_demographics.head(20)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Defining variables

purchase_count = purchase_data["Age Range"].value_counts()



groupby_age_range = purchase_data.groupby(["Age Range"])

unique_SN_age = groupby_age_range["SN"].nunique()

average_purchase_price = groupby_age_range["Price"].mean().round(2)

total_purchase = groupby_age_range["Price"].sum().round(2)

total_per_person = (total_purchase / unique_SN_age).round(2)



purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                    "Average Purchase Price ($)": average_purchase_price,
                                   "Total Purchase Value ($)": total_purchase,
                                   "Avg Total Purchase per Person ($)": total_per_person})

purchasing_analysis.head(10)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# SN_count variable (based on purchase_data, not yet sorted)

SN_count = purchase_data["SN"].value_counts()

# Groupby SN

groupby_SN = purchase_data.groupby(["SN"])

# SN_value and SN_price variable (based on groupby, data needs to already be sorted)

SN_value = groupby_SN["Price"].sum()

SN_price = SN_value / SN_count

# Create dataframe

top_spenders = pd.DataFrame({"Purchase Count": SN_count,
                            "Average Purchase Price": SN_price.round(2),
                            "Total Purchase Value": SN_value})

# Sort Total Purchase Value descending

top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

# Display sorted dataframe

top_spenders.head(5)
```
